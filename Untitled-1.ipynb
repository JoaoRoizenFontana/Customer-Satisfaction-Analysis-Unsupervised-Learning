```python
import pandas as pd

# Carregar o dataset fornecido pelo usuário
file_path = './satisfaction_data.csv'
data = pd.read_csv(file_path)

# Exibir as primeiras linhas do dataset para entender sua estrutura
data.head()
```

O dataset possui várias colunas, incluindo informações sobre passageiros, suas características, tipos de viagem, classe de serviço, pontuações de serviço e satisfação. Antes de iniciar a análise com aprendizado não supervisionado, vou seguir os seguintes passos:

1. **Pré-processamento dos dados:**
   - Remoção de colunas irrelevantes (como identificadores únicos).
   - Tratamento de valores nulos (se houver).
   - Codificação de variáveis categóricas para numéricas.

2. **Análise exploratória dos dados:**
   - Verificar a distribuição das variáveis.
   - Identificar variáveis com maior variância.

3. **Algoritmos de agrupamento:**
   - Testar diferentes algoritmos (como K-Means, DBSCAN e Agglomerative Clustering).
   - Analisar a qualidade dos agrupamentos com métricas como:
     - Inércia.
     - Silhouette.
     - Índice de Calinski-Harabasz.
     - Homogeneidade e completude.
   
4. **Avaliação final:**
   - Comparar os agrupamentos gerados com as classes originais para verificar a correspondência.
   - Explicar os resultados e as possíveis razões para o desempenho.

```python
# Removendo colunas irrelevantes e identificadores únicos
data_cleaned = data.drop(columns=['Unnamed: 0', 'id'])

# Verificando valores nulos e tratando-os, se necessário
missing_values = data_cleaned.isnull().sum()
missing_values_percentage = (missing_values / len(data_cleaned)) * 100

# Exibir os valores nulos para decidir sobre o tratamento
missing_values_percentage
```

Apenas a coluna **"Arrival Delay in Minutes"** possui valores nulos, representando cerca de 0,32% dos dados. Vou tratar esses valores substituindo-os pela mediana, já que atrasos podem ter distribuição assimétrica.

Além disso, é necessário converter as variáveis categóricas em variáveis numéricas para realizar o agrupamento.

```python
# Substituindo valores nulos pela mediana na coluna "Arrival Delay in Minutes"
data_cleaned['Arrival Delay in Minutes'].fillna(data_cleaned['Arrival Delay in Minutes'].median(), inplace=True)

# Convertendo variáveis categóricas em numéricas
categorical_columns = data_cleaned.select_dtypes(include=['object']).columns

data_encoded = pd.get_dummies(data_cleaned, columns=categorical_columns, drop_first=True)

# Verificando a estrutura dos dados após a codificação
data_encoded.info()
```

Os dados foram pré-processados com sucesso:

- A coluna com valores nulos foi tratada substituindo os valores pela mediana.
- Variáveis categóricas foram convertidas em variáveis numéricas por meio de codificação one-hot.
- A coluna "satisfaction_satisfied" representa a classe original e será usada apenas para avaliação posterior.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score
import matplotlib.pyplot as plt

# Removendo a coluna de classe para realizar o agrupamento
X = data_encoded.drop(columns=['satisfaction_satisfied'])

# Testando diferentes números de clusters para K-Means
inertia = []
silhouette_scores = []
calinski_harabasz_scores = []
range_n_clusters = range(2, 11)

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)
    labels = kmeans.labels_
    
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X, labels))
    calinski_harabasz_scores.append(calinski_harabasz_score(X, labels))

# Plotando as métricas para diferentes números de clusters
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(range_n_clusters, inertia, marker='o')
plt.title("Inércia")
plt.xlabel("Número de Clusters")
plt.ylabel("Inércia")

plt.subplot(1, 3, 2)
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.title("Coeficiente de Silhouette")
plt.xlabel("Número de Clusters")
plt.ylabel("Silhouette Score")

plt.subplot(1, 3, 3)
plt.plot(range_n_clusters, calinski_harabasz_scores, marker='o')
plt.title("Índice de Calinski-Harabasz")
plt.xlabel("Número de Clusters")
plt.ylabel("Calinski-Harabasz Score")

plt.tight_layout()
plt.show()
```

O gráfico representa a análise do desempenho do algoritmo K-Means para diferentes números de clusters, utilizando três métricas principais:

1. **Inércia:** (Gráfico da esquerda)  
   - Mede a soma das distâncias quadradas entre os pontos e seus centróides.  
   - Um valor menor indica clusters mais compactos.  
   - No entanto, há um ponto de "diminuição marginal" (elbow point), onde a redução de inércia começa a ser menos significativa ao aumentar o número de clusters. Isso geralmente indica o número ideal de clusters. No gráfico, isso parece ocorrer em torno de **3 ou 4 clusters**.

2. **Coeficiente de Silhouette:** (Gráfico do meio)  
   - Mede o quão bem cada ponto está dentro do seu cluster e distante de outros clusters.  
   - O valor varia de -1 a 1, sendo valores mais altos melhores.  
   - O gráfico mostra que o valor é mais alto para **2 clusters**, mas cai gradualmente ao aumentar o número de clusters, indicando que a separação entre clusters diminui com mais divisões.

3. **Índice de Calinski-Harabasz:** (Gráfico da direita)  
   - Avalia a proporção da dispersão total entre os clusters em relação à dispersão dentro dos clusters.  
   - Valores maiores indicam melhores separações entre clusters.  
   - O gráfico apresenta um crescimento constante, indicando que mais clusters resultam em maior separação, embora isso possa ser devido ao aumento de complexidade.

### Conclusão:
- A escolha do número de clusters deve equilibrar as três métricas.  
- Com base no gráfico, **3 ou 4 clusters** parecem ser o número ideal, já que representam um bom equilíbrio entre inércia baixa e um Coeficiente de Silhouette relativamente alto.  
- A decisão final pode ser ajustada com base em requisitos específicos do problema ou em experimentos adicionais.

```python
from sklearn.metrics import homogeneity_score, completeness_score, adjusted_rand_score, accuracy_score
import numpy as np
```

```python
# Selecionando o número ideal de clusters com base nas métricas
optimal_clusters = silhouette_scores.index(max(silhouette_scores)) + 2 

# Aplicando K-Means com o número ideal de clusters
kmeans_optimal = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans_optimal.fit(X)
labels_optimal = kmeans_optimal.labels_

# Avaliando a correspondência dos clusters com as classes originais
homogeneity = homogeneity_score(data_encoded['satisfaction_satisfied'], labels_optimal)
completeness = completeness_score(data_encoded['satisfaction_satisfied'], labels_optimal)
adjusted_rand = adjusted_rand_score(data_encoded['satisfaction_satisfied'], labels_optimal)

homogeneity, completeness, adjusted_rand
```

```python
# Aplicando K-Means com o número ideal de clusters identificado (4 clusters neste caso)
optimal_clusters = 4
kmeans_final = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans_final.fit(X)
final_labels = kmeans_final.labels_

# Avaliando os agrupamentos com as métricas de homogeneidade, completude e ARI
homogeneity = homogeneity_score(data_encoded['satisfaction_satisfied'], final_labels)
completeness = completeness_score(data_encoded['satisfaction_satisfied'], final_labels)
adjusted_rand = adjusted_rand_score(data_encoded['satisfaction_satisfied'], final_labels)

# Calculando a taxa de acerto com base na correspondência com a classe original
# Ajustando os rótulos para comparar diretamente com as classes originais (caso necessário)
# Mapeando clusters para classes reais
mapping = {cluster: np.argmax(np.bincount(data_encoded['satisfaction_satisfied'][final_labels == cluster])) 
           for cluster in range(optimal_clusters)}
mapped_labels = np.array([mapping[label] for label in final_labels])

# Calculando a taxa de acerto
accuracy = accuracy_score(data_encoded['satisfaction_satisfied'], mapped_labels)

homogeneity, completeness, adjusted_rand, accuracy
```

Com base nos resultados:

- **Homogeneidade (0.068):** Os clusters não são homogêneos, indicando que há uma grande mistura de classes nos agrupamentos.
- **Completude (0.038):** Mostra que pontos da mesma classe estão espalhados entre diferentes clusters, indicando baixa completude.
- **Índice Rand Ajustado (ARI, 0.054):** O valor muito baixo sugere que os agrupamentos não estão bem alinhados com as classes reais, mesmo considerando ajustes para aleatoriedade.
- **Taxa de Acerto (65.8%):** Apesar de um ARI baixo, a taxa de acerto é moderada, indicando que os clusters possuem certa correspondência com as classes, mas com erros consideráveis.

### Próximos Passos:

1. **Testar outros algoritmos de agrupamento:**
   - Implementar **DBSCAN**, que pode lidar melhor com ruídos e encontrar clusters de formas arbitrárias.
   - Implementar **Agglomerative Clustering** (agrupamento hierárquico) para verificar se ele oferece uma melhor estruturação dos dados.

2. **Validar clusters adicionais:**
   - Calcular as métricas de Silhouette e Calinski-Harabasz para os novos algoritmos.

3. **Analisar features:**
   - Identificar as variáveis mais importantes para a separação entre classes (análise de importância ou PCA).

```python
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering

# DBSCAN
dbscan = DBSCAN(eps=3, min_samples=5)  # Ajuste de parâmetros necessários
dbscan_labels = dbscan.fit_predict(X)

# Avaliação do DBSCAN
dbscan_homogeneity = homogeneity_score(data_encoded['satisfaction_satisfied'], dbscan_labels)
dbscan_completeness = completeness_score(data_encoded['satisfaction_satisfied'], dbscan_labels)
dbscan_ari = adjusted_rand_score(data_encoded['satisfaction_satisfied'], dbscan_labels)

# Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=4)  # Número de clusters identificado
agg_labels = agg_clustering.fit_predict(X)

# Avaliação do Agglomerative Clustering
agg_homogeneity = homogeneity_score(data_encoded['satisfaction_satisfied'], agg_labels)
agg_completeness = completeness_score(data_encoded['satisfaction_satisfied'], agg_labels)
agg_ari = adjusted_rand_score(data_encoded['satisfaction_satisfied'], agg_labels)

# Retorne os resultados para cada algoritmo
{
    "DBSCAN": {
        "homogeneity": dbscan_homogeneity,
        "completeness": dbscan_completeness,
        "ari": dbscan_ari
    },
    "Agglomerative": {
        "homogeneity": agg_homogeneity,
        "completeness": agg_completeness,
        "ari": agg_ari
    }
}
```

### **DBSCAN:**
- **Homogeneidade (0.0):** Nenhum cluster contém apenas pontos de uma única classe. Isso ocorre porque o DBSCAN provavelmente considerou a maioria dos pontos como ruído ou agrupou tudo em um único cluster.
- **Completude (1.0):** Todos os pontos de uma mesma classe foram atribuídos ao mesmo cluster, mas isso é enganoso, já que, na prática, houve problemas no agrupamento.
- **ARI (0.0):** Total ausência de alinhamento entre os clusters gerados e as classes reais.

**Conclusão:** DBSCAN não conseguiu identificar clusters significativos. Isso pode ser devido à alta dimensionalidade dos dados ou à necessidade de ajustar os parâmetros `eps` e `min_samples`.

---

### **Agglomerative Clustering:**
- **Homogeneidade (0.062):** Baixa homogeneidade indica que os clusters contêm uma grande mistura de classes.
- **Completude (0.035):** Poucos elementos da mesma classe foram atribuídos ao mesmo cluster.
- **ARI (0.046):** A baixa correspondência entre os agrupamentos e as classes reais sugere que o método hierárquico também não é ideal para este conjunto de dados.

**Conclusão:** Embora o Agglomerative Clustering tenha desempenho ligeiramente melhor do que DBSCAN, ele ainda não conseguiu separar bem as classes.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Aplicando PCA para redução para 2 componentes principais
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualizando os dados reduzidos
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data_encoded['satisfaction_satisfied'], cmap='viridis', s=10)
plt.title("Visualização dos Dados Reduzidos (PCA)")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.colorbar(label="Classe Real")
plt.show()

# Aplicar DBSCAN novamente com dados reduzidos
dbscan_reduced = DBSCAN(eps=0.5, min_samples=5)  # Ajustar parâmetros conforme necessário
dbscan_reduced_labels = dbscan_reduced.fit_predict(X_pca)

# Avaliar clusters DBSCAN após PCA
dbscan_reduced_homogeneity = homogeneity_score(data_encoded['satisfaction_satisfied'], dbscan_reduced_labels)
dbscan_reduced_completeness = completeness_score(data_encoded['satisfaction_satisfied'], dbscan_reduced_labels)
dbscan_reduced_ari = adjusted_rand_score(data_encoded['satisfaction_satisfied'], dbscan_reduced_labels)

{
    "DBSCAN (PCA)": {
        "homogeneity": dbscan_reduced_homogeneity,
        "completeness": dbscan_reduced_completeness,
        "ari": dbscan_reduced_ari
    }
}
```

O gráfico da **redução de dimensionalidade (PCA)** mostra os dados projetados em duas dimensões principais, coloridos de acordo com as classes reais. Isso nos ajuda a entender visualmente a separação entre as classes.

### Observações do Gráfico:
1. **Sobreposição de Classes:**
   - Há uma sobreposição considerável entre os grupos, indicando que as classes podem não ser bem separáveis no espaço reduzido.

2. **Distribuição de Pontos:**
   - A maioria dos pontos está concentrada em regiões próximas, enquanto alguns pontos estão espalhados em áreas menos densas.

3. **Impacto no Agrupamento:**
   - Essa sobreposição pode dificultar a tarefa de agrupamento, resultando em baixos valores de homogeneidade e completude, como vimos anteriormente.

### Próximos Passos:
Agora que visualizamos os dados reduzidos, podemos usar essa projeção para:

1. **Avaliar a Qualidade do DBSCAN:**
   - Verificar se a redução de dimensionalidade melhora os agrupamentos gerados.

2. **Explorar a Estabilidade de Clusters:**
   - Recalcular as métricas de Silhouette e Calinski-Harabasz usando os dados reduzidos.



```python
# Aplicar DBSCAN após PCA
dbscan_reduced = DBSCAN(eps=50, min_samples=10)  # Ajuste de eps baseado na densidade percebida
dbscan_reduced_labels = dbscan_reduced.fit_predict(X_pca)

# Avaliação dos clusters DBSCAN com dados reduzidos
dbscan_reduced_homogeneity = homogeneity_score(data_encoded['satisfaction_satisfied'], dbscan_reduced_labels)
dbscan_reduced_completeness = completeness_score(data_encoded['satisfaction_satisfied'], dbscan_reduced_labels)
dbscan_reduced_ari = adjusted_rand_score(data_encoded['satisfaction_satisfied'], dbscan_reduced_labels)

# Calculando o coeficiente de Silhouette e o índice de Calinski-Harabasz
if len(set(dbscan_reduced_labels)) > 1:  # Evitar erro quando há apenas 1 cluster
    dbscan_reduced_silhouette = silhouette_score(X_pca, dbscan_reduced_labels)
    dbscan_reduced_calinski = calinski_harabasz_score(X_pca, dbscan_reduced_labels)
else:
    dbscan_reduced_silhouette = None
    dbscan_reduced_calinski = None

{
    "DBSCAN (PCA)": {
        "homogeneity": dbscan_reduced_homogeneity,
        "completeness": dbscan_reduced_completeness,
        "ari": dbscan_reduced_ari,
        "silhouette": dbscan_reduced_silhouette,
        "calinski_harabasz": dbscan_reduced_calinski
    }
}
```
